```python
import pandas as pd
df = pd.read_csv("../inputs/runs.csv")
```

```python
no_converged_matrices = []
for name, g in df.groupby('matrix_name'):
    if not g['is_converged'].astype(int).astype(bool).any():
        no_converged_matrices.append(name)
        print(name)

print(f"未収束の行列数: {len(no_converged_matrices)}")
```

```python
# is_converged をブール化（NaNは False 扱い）
is_conv = (
    df['is_converged']
      .fillna(0)
      .astype(int)        # 0/1, True/False, '0'/'1' を吸収
      .astype(bool)
)

# 行列ごとに「どれか一つでも収束したか」を集計
never_converged_names = (
    df.assign(is_conv=is_conv)
      .groupby('matrix_name', sort=False)['is_conv']
      .any()              # True: どれかの前処理で収束
      .pipe(lambda s: s[~s])  # False だけ抽出
      .index
      .tolist()
)

for name in never_converged_names:
    print(name)
```
