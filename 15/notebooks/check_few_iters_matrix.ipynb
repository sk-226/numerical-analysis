```python
import pandas as pd
df = pd.read_csv("runs.csv")
```

```python
n_iters = 3
few_iters_df = df.query("(iters <= @n_iters) & (iters > 0)")
few_iters_df
```

```python
few_iters_df["prec_label"].value_counts()
```

```python
few_iters_matrices = few_iters_df["matrix_name"].unique()
print(few_iters_matrices)
print(f"COUNT: {len(few_iters_matrices)}")
```

```python
few_iters_df
```

```python
ban = few_iters_matrices

# 行列名で除外（該当する行を全部消す）
df_filtered = df.query("matrix_name not in @ban").copy()

# 必要なら上書き保存
df_filtered.to_csv("results.filtered.csv", index=False)
```
